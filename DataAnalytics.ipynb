```python
#Q1: tablet with more than 3 units sold
df.query("Product=='Tablet' and `Units Sold`>3")
```

```python
#Q2:Sales in South with revenue more than 5000
df.query("Region=='South' and  Revenue>5000")
```

```python
#Q3: Show transactions where product is mouse or units sold > 5 or region is west
df.query("Product=='Mouse' or `Units Sold`>5 or Region == 'West'")
```

```python
#Q4: show transactions of accessories only
df.query("Category=='Accessories'")
```

```python
#Q5:Show transactions where price per unit is < 500
df.query("`Price Per Unit` < 500")
```

```python
#Q6:Using df.loc()
df.loc[(df['Product']=='Mouse') & (df['Region']=='South'),['OrderID', 'Region', 'Product']]
```
