**LINE CHART**

```python
import matplotlib.pyplot as plt
```

```python
#data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
Sales_A = [30, 34, 38, 42, 46, 52]
Sales_B = [30, 35, 40, 42, 48, 50]
```

```python
#Create Line Chart for Product A
plt.plot(months, Sales_A, label='Product A', color='red', marker='o', linestyle='-')

#Create Line Chart for Product B
plt.plot(months, Sales_B, label='Product B', color='blue', marker='o', linestyle='-')

#Add titles and labels
plt.title('Monthly Sales Trends')
plt.xlabel('Months')
plt.ylabel('Sales (rs. 1000)')

#Ad legend and grid
plt.legend()
plt.grid(True)

#Display the chart
plt.show()

```

**SCATTER PLOT**

```python
import matplotlib.pyplot as plt
```

```python
#Sample data
temperature = [20, 22, 24, 26, 28, 30, 32, 34, 36, 38]
sales = [10, 12, 15, 18, 22, 26, 30, 35, 40, 45]

#Create Scatter Plot
plt.scatter(temperature, sales, color='green', marker='o')

#Add titles and labels
plt.title('Sales vs Temperature')
plt.xlabel('Temperature')
plt.ylabel('Sales')

#Show grid and plot
plt.grid(True)
plt.show()
```
